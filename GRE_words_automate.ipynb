```python
def get_word_info(word):
    prompt = f"Word: {word}\nMeaning: "
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=50,
        temperature=0.5,
        stop=None
    )
    meaning = response['choices'][0]['text'].strip()

    prompt = f"Word: {word}\nSynonyms: "
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=50,
        temperature=0.5,
        stop=None
    )
    synonyms = response['choices'][0]['text'].strip()

    prompt = f"Word: {word}\nAntonyms: "
    response = openai.Completion.create(
        engine="text-davinci-003",
        prompt=prompt,
        max_tokens=50,
        temperature=0.5,
        stop=None
    )
    antonyms = response['choices'][0]['text'].strip()

    return meaning, synonyms, antonyms

# List of words to process

# Create a DataFrame
columns = ["Word", "Meaning", "Synonyms", "Antonyms"]
df = pd.DataFrame(columns=columns)

# Process each word and append to the DataFrame
for word in word_list:
    meaning, synonyms, antonyms = get_word_info(word)
    df = df.append({"Word": word, "Meaning": meaning, "Synonyms": synonyms, "Antonyms": antonyms}, ignore_index=True)

# Save the DataFrame to a CSV file
output_csv_path = 'gre.csv'
df.to_csv(output_csv_path, index=False)

print(f"Data has been saved to {output_csv_path}")
```
